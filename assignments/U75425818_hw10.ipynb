```python
#Problem 1:
#plot the probability of picking k consecutive green balls

#Import libraries
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

#Define variables
n = 10
p = 0.5
k = np.arange(0,11)

#Define the binomial distribution
binomial = stats.binom.pmf(k, n, p)

#Plot the binomial distribution
plt.plot(k, binomial, 'o-')
plt.title('Binomial: n = %i , p = %.2f' % (n,p), fontsize=15)
plt.xlabel('Number of green balls')
plt.ylabel('Probability of green balls')
plt.show()
```

```python
#Problem 3:
#Esimate π with Monte Carlo simulation

#Import libraries
import numpy as np
import random

#1) n = 10
#Define variables
n = 10
x = np.random.uniform(-1,1,n)
y = np.random.uniform(-1,1,n)

#Define the circle
circle = np.sqrt(x**2 + y**2)
inside = circle[circle <= 1]
outside = circle[circle > 1]

pi_r = len(inside) / len(outside)

#print pi 
print("Pi with n = 10 => ", pi_r)

#2) n = 20
#Define variables
n = 20
x = np.random.uniform(-1,1,n)
y = np.random.uniform(-1,1,n)

#Define the circle
circle = np.sqrt(x**2 + y**2)
inside = circle[circle <= 1]
outside = circle[circle > 1]

pi_r = len(inside) / len(outside)

#print pi 
print("Pi with n = 20 => ", pi_r)

#3) absolute error |4n/N − π < 10^−5?
#Define variables
pi = 3.1315
err = 0.00001
min_err = 0.00002
n = 0
circle = 1
square = 1
#Define the circle
while min_err > err:
    n = n + 1
    x = random.uniform(-1,1)
    y = random.uniform(-1,1)

    dist = x**2 + y**2

    if dist <= 1:
        circle = circle + 1
    square = square + 1
    
    pi_r = 4*circle / square
    min_err = abs(pi_r - pi)
    print(min_err)

print("Pi with n = ", n, " => ", pi_r)



```
