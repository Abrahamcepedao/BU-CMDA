```python
#Problem 1. 
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace (1 , 5 , 100) # x value - continuous
n = np.arange (1 , 6 , 1) # n values (1,2,3,4,5)

#1. f(x) = 1 --> x
def f1(x):
    return x

height = [f1(1* i) for i in range (1,6)] #height values

plt.plot(x, f1(x), color="green")
plt.title("1. $f(x) = $x")
plt.bar(n, height, width = 1,alpha = 0.9, color="red")
plt.show()

#2. f(x) = x --> (x^2)/2
def f2(x):
    return (x*x)/2

height = [f2(1* i) for i in range (1,6)] #height values

plt.plot(x, f2(x), color="green")
plt.title("2. $f(x) = $(x^2)/2")
plt.bar(n, height, width = 1,alpha = 0.9, color="red")
plt.show()


#3. f(x) = x^2 --> (x^3)/3
def f3(x):
    return pow(x,3)/3

height = [f3(1* i) for i in range (1,6)] #height values

plt.plot(x, f3(x), color="green")
plt.title("3. $f(x) = $(x^3)/3")
plt.bar(n, height, width = 1,alpha = 0.9, color="red")
plt.show()


#4. f(x) = x^3 --> (x^4)/4
def f4(x):
    return pow(x,4)/4

height = [f4(1* i) for i in range (1,6)] #height values

plt.plot(x, f4(x), color="green")
plt.title("4. $f(x) = $(x^4)/4")
plt.bar(n, height, width = 1,alpha = 0.9, color="red")
plt.show()


#5. f(x) = âˆšx --> (2/3)*x^(3/2)
def f5(x):
    return (2/3)*pow(x, 3/2)

height = [f5(1* i) for i in range (1,6)] #height values

plt.plot(x, f5(x), color="green")
plt.title("5. $f(x) = $(2/3)*x^(3/2)")
plt.bar(n, height, width = 1,alpha = 0.9, color="red")
plt.show()

#6. f(x) = e^x --> e^x
def f6(x):
    return np.exp(x)

height = [f6(1* i) for i in range (1,6)] #height values

plt.plot(x, f6(x), color="green")
plt.title("6. $f(x) = $e^x")
plt.bar(n, height, width = 1,alpha = 0.9, color="red")
plt.show()

#7. f(x) = 2^x --> 2^x/(log(2))
def f7(x):
    return pow(2,x)/np.log(2)

height = [f7(1* i) for i in range (1,6)] #height values

plt.plot(x, f7(x), color="green")
plt.title("7. $f(x) = $2^x/(log(2))")
plt.bar(n, height, width = 1,alpha = 0.9, color="red")
plt.show()

#8. f(x) = log(x) --> xlog(x)-x
def f8(x):
    return x*np.log(x)-x

height = [f8(1* i) for i in range (1,6)] #height values

plt.plot(x, f8(x), color="green")
plt.title("8. $f(x) = $xlog(x)-x")
plt.bar(n, height, width = 1,alpha = 0.9, color="red")
plt.show()


#9. f(x) = sin(x) --> -cos(x)
def f9(x):
    return -1*np.cos(x)

height = [f9(1* i) for i in range (1,6)] #height values

plt.plot(x, f9(x), color="green")
plt.title("9. $f(x) = $-cos(x)")
plt.bar(n, height, width = 1,alpha = 0.9, color="red")
plt.show()

#10. f(x) = cos(x) --> sin(x)
def f10(x):
    return np.sin(x)

height = [f10(1* i) for i in range (1,6)] #height values

plt.plot(x, f10(x), color="green")
plt.title("10. $f(x) = $sin(x)")
plt.bar(n, height, width = 1,alpha = 0.9, color="red")
plt.show()


```

```python
#Problem 4 - Compute integral with fixed quadrature, simpson, and trapezoid from 1 to 5
import scipy . integrate as spi
import numpy as np

a, b= 1 , 5



###1. f(x) = 1
integrand = lambda x : x-x+1

#fixed quadrature
res, err = spi.fixed_quad(integrand,a,b)
print ("1. quadrature  = ", round(res,3))

#trapezoid
step = 1e-4
xs = np.arange(a, b, step)
ys = integrand(xs)
res = spi.trapezoid(ys , xs)
print ("1. simpson     = ", round(res,3))

#simpson
res = spi.simpson(ys , xs)
print ("1. trapezoid   = ", round(res,3))


###2. f(x) = x
print("--------")
integrand = lambda x: x

#fixed quadrature
res, err = spi.quad(integrand,a,b)
print ("2. quadrature  = ", round(res,3))

#trapezoid
step = 1e-4
xs = np.arange(a, b, step)
ys = integrand(xs)
res = spi.trapezoid(ys , xs)
print ("2. simpson     = ", round(res,3))

#simpson
res = spi.simpson(ys , xs)
print ("2. trapezoid   = ", round(res,3))


#3. f(x) = x^2
print("--------")
integrand = lambda x: x*x

#fixed quadrature
res, err = spi.quad(integrand,a,b)
print ("3. quadrature  = ", round(res,3))

#trapezoid
step = 1e-4
xs = np.arange(a, b, step)
ys = integrand(xs)
res = spi.trapezoid(ys , xs)
print ("3. simpson     = ", round(res,3))

#simpson
res = spi.simpson(ys , xs)
print ("3. trapezoid   = ", round(res,3))


#4. f(x) = x^3
print("--------")
integrand = lambda x: pow(x,3)

#fixed quadrature
res, err = spi.quad(integrand,a,b)
print ("4. quadrature  = ", round(res,3))

#trapezoid
step = 1e-4
xs = np.arange(a, b, step)
ys = integrand(xs)
res = spi.trapezoid(ys , xs)
print ("4. simpson     = ", round(res,3))

#simpson
res = spi.simpson(ys , xs)
print ("4. trapezoid   = ", round(res,3))


#5. f(x) = âˆšx
print("--------")
integrand = lambda x: np.sqrt(x)

#fixed quadrature
res, err = spi.quad(integrand,a,b)
print ("5. quadrature  = ", round(res,3))

#trapezoid
step = 1e-4
xs = np.arange(a, b, step)
ys = integrand(xs)
res = spi.trapezoid(ys , xs)
print ("5. simpson     = ", round(res,3))

#simpson
res = spi.simpson(ys , xs)
print ("5. trapezoid   = ", round(res,3))

#6. f(x) = e^x
print("--------")
integrand = lambda x: np.exp(x)

#fixed quadrature
res, err = spi.quad(integrand,a,b)
print ("6. quadrature  = ", round(res,3))

#trapezoid
step = 1e-4
xs = np.arange(a, b, step)
ys = integrand(xs)
res = spi.trapezoid(ys , xs)
print ("6. simpson     = ", round(res,3))

#simpson
res = spi.simpson(ys , xs)
print ("6. trapezoid   = ", round(res,3))


#7. f(x) = 2^x
print("--------")
integrand = lambda x: pow(2,x)

#fixed quadrature
res, err = spi.quad(integrand,a,b)
print ("7. quadrature  = ", round(res,3))

#trapezoid
step = 1e-4
xs = np.arange(a, b, step)
ys = integrand(xs)
res = spi.trapezoid(ys , xs)
print ("7. simpson     = ", round(res,3))

#simpson
res = spi.simpson(ys , xs)
print ("7. trapezoid   = ", round(res,3))


#8. f(x) = log(x)
print("--------")
integrand = lambda x: np.log(x)

#fixed quadrature
res, err = spi.quad(integrand,a,b)
print ("8. quadrature  = ", round(res,3))

#trapezoid
step = 1e-4
xs = np.arange(a, b, step)
ys = integrand(xs)
res = spi.trapezoid(ys , xs)
print ("8. simpson     = ", round(res,3))

#simpson
res = spi.simpson(ys , xs)
print ("8. trapezoid   = ", round(res,3))


#9. f(x) = sin(x)
print("--------")
integrand = lambda x: np.sin(x)

#fixed quadrature
res, err = spi.quad(integrand,a,b)
print ("9. quadrature  = ", round(res,3))

#trapezoid
step = 1e-4
xs = np.arange(a, b, step)
ys = integrand(xs)
res = spi.trapezoid(ys , xs)
print ("9. simpson     = ", round(res,3))

#simpson
res = spi.simpson(ys , xs)
print ("9. trapezoid   = ", round(res,3))


#10. f(x) = cos(x)
print("--------")
integrand = lambda x: np.cos(x)

#fixed quadrature
res, err = spi.quad(integrand,a,b)
print ("10. quadrature  = ", round(res,3))

#trapezoid
step = 1e-4
xs = np.arange(a, b, step)
ys = integrand(xs)
res = spi.trapezoid(ys , xs)
print ("10. simpson     = ", round(res,3))

#simpson
res = spi.simpson(ys , xs)
print ("10. trapezoid   = ", round(res,3))
```

```python
##Problem 5 - Compute integral numerically
import numpy as np

a = [1,2,3,4]
b = [2,3,4,5]
c_sum = 0

#Quadrature function
def quad(f, i):
    return (f(a[i]) + f(b[i])) / 2

#Trapezoid function
def trap(f, i):
    return f((a[i] + b[i]) / 2)

#Simpson function
def simp(f, i):
    return (f(a[i]) + 4 * f((a[i] + b[i]) / 2) + f(b[i])) / 6



#1. f(x) = 1
def f1(x):
    return x-x+1


#2. f(x) = x
def f2(x):
    return x


#3. f(x) = x^2
def f3(x):
    return x*x

#4. f(x) = x^3
def f4(x):
    return pow(x,3)


#5. f(x) = âˆšx
def f5(x):
    return np.sqrt(x)


#6. f(x) = e^x
def f6(x):
    return np.exp(x)


#7. f(x) = 2^x
def f7(x):
    return pow(2,x)


#8. f(x) = log(x)
def f8(x):
    return np.log(x)


#9. f(x) = sin(x)
def f9(x):
    return np.sin(x)

#10. f(x) = cos(x)
def f10(x):
    return np.cos(x)


fs = [f1,f2,f3,f4,f5,f6,f7,f8,f9,f10]

for e in range(10):
    c_sum = 0
    print("\n--------\n")
    
    
    print(e+1, ". F(x)")
    for i in range(4):
        res = fs[e](i+1)
        print(round(res,2))
    
    print(e+1, ". Trapezoid")
    for i in range(4):
        res = trap(fs[e], i)
        c_sum += res
        print(round(res,2))

    print("Tot = ", round(c_sum,2))
    c_sum = 0

    
    print(e+1, ". Quadrature")
    for i in range(4):
        res = quad(fs[e], i)
        c_sum += res
        print(round(res,2))

    
    print("Tot = ", round(c_sum,2))
    c_sum = 0
    

    print(e+1, ". Simpson")
    for i in range(4):
        res = simp(fs[e], i)
        c_sum += res
        print(round(res,2))
        
    print("Tot = ", round(c_sum,2))
    c_sum = 0


```

```python
##Problem 6 - Calculate relative error
import numpy as np

a = [1,2,3,4]
b = [2,3,4,5]
c_sum = 0

#Quadrature function
def quad(f, i):
    return (f(a[i]) + f(b[i])) / 2

#Trapezoid function
def trap(f, i):
    return f((a[i] + b[i]) / 2)

#Simpson function
def simp(f, i):
    return (f(a[i]) + 4 * f((a[i] + b[i]) / 2) + f(b[i])) / 6



#1. f(x) = 1
def f1(x):
    return x-x+1


#2. f(x) = x
def f2(x):
    return x


#3. f(x) = x^2
def f3(x):
    return x*x

#4. f(x) = x^3
def f4(x):
    return pow(x,3)


#5. f(x) = âˆšx
def f5(x):
    return np.sqrt(x)


#6. f(x) = e^x
def f6(x):
    return np.exp(x)


#7. f(x) = 2^x
def f7(x):
    return pow(2,x)


#8. f(x) = log(x)
def f8(x):
    return np.log(x)


#9. f(x) = sin(x)
def f9(x):
    return np.sin(x)

#10. f(x) = cos(x)
def f10(x):
    return np.cos(x)


fs = [f1,f2,f3,f4,f5,f6,f7,f8,f9,f10]


import sympy as sym


x = sym.Symbol('x')


real_values = [sym.integrate(1, (x, 1, 5)), 
               sym.integrate(x, (x, 1, 5)), 
               sym.integrate(x*x, (x, 1, 5)), 
               sym.integrate(pow(x,3), (x, 1, 5)),
               sym.integrate(sym.sqrt(x), (x, 1, 5)),
               sym.integrate(sym.exp(x), (x, 1, 5)),
               sym.integrate(pow(2,x), (x, 1, 5)),
               sym.integrate(sym.log(x), (x, 1, 5)),
               sym.integrate(sym.sin(x), (x, 1, 5)),
               sym.integrate(sym.cos(x), (x, 1, 5))
              ]

for e in range(10):
    c_sum = 0
    print("\n--------\n")

    real = real_values[e]
    
    print(e+1, ". Trapezoid")
    for i in range(4):
        res = trap(fs[e], i)
        c_sum += res

    print("Err = ", round( (c_sum - real)/real*100, 2) )
    c_sum = 0
    
    print(e+1, ". Quadrature")
    for i in range(4):
        res = quad(fs[e], i)
        c_sum += res

    
    print("Err = ", round((c_sum - real)/real*100, 2) )
    c_sum = 0
    

    print(e+1, ". Simpson")
    for i in range(4):
        res = simp(fs[e], i)
        c_sum += res
        
    print("Err = ", round((c_sum - real)/real*100, 2) )
    c_sum = 0
```
