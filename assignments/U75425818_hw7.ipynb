```python
import numpy as np
from numpy.linalg import det
import matplotlib.pyplot as plt


def Circle(x,y):
    return (x*x+y*y)

xx=np.linspace(-1,1,1000)
yy=np.linspace(-1,1,1000)
[X,Y]=np.meshgrid(xx,yy)


#1.1 M1 - expand x by a factor of 1.5
M1 = np.array([[1.5,0],[0,1]])
M1_val = np.pi * det(M1)
print("M1 = ", M1_val)

#1.2 M2 - contract y by a factor of 0.75
M2 = np.array([[1,0],[0,0.75]])
M2_val = np.pi * det(M2)
print("M2 = ", M2_val)

#1.3 M3 - shear along x by a factor of 2
M3 = np.array([[1,2],[0,1]])
M3_val = np.pi * det(M3)
print("M3 = ", M3_val)

#1.4 M4 - shear along y by a factor of 1.25
M4 = np.array([[1,0],[1.25,1]])
M4_val = np.pi * det(M4)
print("M4 = ", M4_val)

#1.5 M5 - counterclockwise rotation by pi/4
theta = np.pi/4
M5 = np.array([[np.cos(theta),-np.sin(theta)],[np.sin(theta),np.cos(theta)]])
M5_val = np.pi * det(M5)
print("M5 = ", M5_val)

#1.6 M6 - clockwise rotation by pi/6
theta2 = np.pi/6
M6 = np.array([[np.cos(theta2),np.sin(theta2)],[-np.sin(theta2),np.cos(theta2)]])
M6_val = np.pi * det(M6)
print("M6 = ", M6_val)

 
figure, axis = plt.subplots(3, 2, figsize=(20,20))

axis[0][0].set_aspect( 1 )
axis[0][1].set_aspect( 1 )
axis[1][0].set_aspect( 1 )
axis[1][1].set_aspect( 1 )
axis[2][0].set_aspect( 1 )
axis[2][1].set_aspect( 1 )


#M1
Z = Circle(X*M1_val,Y) 
axis[0][0].contour(X,Y,Z,[1])
axis[0][0].set_title("M1")

#M2
Z = Circle(X,Y*M2_val) 
axis[0][1].contour(X,Y,Z,[1])
axis[0][1].set_title("M2")

#M3
Z = Circle(X*M3_val,Y) 
axis[1][0].contour(X,Y,Z,[1])
axis[1][0].set_title("M3")

#M4
Z = Circle(X,Y*M4_val) 
axis[1][1].contour(X,Y,Z,[1])
axis[1][1].set_title("M4")

#M5
Z = Circle(X,Y) 
axis[2][0].contour(X+M5_val,Y+M5_val,Z,[1])
axis[2][0].set_title("M5")

#M6
Z = Circle(X,Y) 
axis[2][1].contour(X-M6_val,Y-M6_val,Z,[1])
axis[2][1].set_title("M6")

plt.show()


```

```python
#Problem 4. Determinant and L1
from numpy.linalg import inv
from numpy.linalg import matrix_power
from numpy.linalg import det
import matplotlib.pyplot as plt


def Circle(x,y):
    return (x*x+y*y)

xx=np.linspace(-1,1,1000)
yy=np.linspace(-1,1,1000)
[X,Y]=np.meshgrid(xx,yy)

#1.1 M1 - expand x by a factor of 1.5
M1 = np.array([[1.5,0],[0,1]])

#1.2 M2 - contract y by a factor of 0.75
M2 = np.array([[1,0],[0,0.75]])

#1.3 M3 - shear along x by a factor of 2
M3 = np.array([[1,2],[0,1]])

#1.4 M4 - shear along y by a factor of 1.25
M4 = np.array([[1,0],[1.25,1]])

#1.5 M5 - counterclockwise rotation by pi/4
theta = np.pi/4
M5 = np.array([[np.cos(theta),-np.sin(theta)],[np.sin(theta),np.cos(theta)]])

#1.6 M6 - clockwise rotation by pi/6
theta2 = np.pi/6
M6 = np.array([[np.cos(theta2),np.sin(theta2)],[-np.sin(theta2),np.cos(theta2)]])

M1_2 = matrix_power(M1,2)
M3_3 = matrix_power(M3,3)
M_1 = M1_2.dot(M3_3)
M_2 = M_1.dot(M5)
MF_1 = np.pi * det(M_2)

M2_I = inv(M2)
M4_I = inv(M4)
M4_I2 = matrix_power(M4_I,2)
M_3 = M2_I.dot(M4_I2)
M_4 = M_1.dot(M6)
MF_2 = np.pi * det(M_4)

figure, axis = plt.subplots(1, 2, figsize=(20,20))

axis[0].set_aspect( 1 )
axis[1].set_aspect( 1 )


#M1
Z = Circle(X*MF_1,Y) 
axis[0].contour(X,Y,Z,[1])
axis[0].set_title("M1")

#M2
Z = Circle(X*MF_2,Y) 
axis[1].contour(X,Y,Z,[1])
axis[1].set_title("M2")

plt.show()
```

```python
#Problem 5. Invert Matrices
import numpy as np
from numpy.linalg import eig
import matplotlib.pyplot as plt

A = np.array([[1,4],[2,3]])
B = np.array([[5,19],[-2,5]])
C = np.array([[2,1,9],[5,7,1],[4,7,1]])

w1,v1 = eig(A)
w2,v2 = eig(B)
w3,v3 = eig(C)

print("A eigenvalues = ", w1)
print("A eigenvectors = ", v1)
print("B eigenvalues = ", w2)
print("B eigenvectors = ", v2)
print("C eigenvalues = ", w3)
print("C eigenvectors = ", v3)

M = [A,B,C]
T = ["A","B","C"]

figure, axis = plt.subplots(3, 1, figsize=(10,10))

#By row
for n in range(len(M)):
    max_sum = 0
    max_rad = 0
    for i in range(len(M[n])):
        c_sum = 0
        for j in range(len(M[n])):
            if j != i:
                c_sum += abs(M[n][i][j])
        rad = M[n][i][i]
        
        #plot circle
        axis[n].set_aspect( 1 )
        axis[n].add_artist( plt.Circle(( c_sum, 0 ), rad, fill = False  ))
        axis[n].plot(c_sum, 0, 'ro')
        
        #set max values for limits
        if c_sum > max_sum:
            max_sum = c_sum
        if rad > max_rad:
            max_rad = rad
        
    axis[n].set_xlim(-max_sum-max_rad,max_sum+max_rad)
    axis[n].set_ylim(-max_rad,max_rad)
    axis[n].set_title(T[n]+" by row")
plt.show()


figure, axis = plt.subplots(3, 1, figsize=(10,10))

#By column
for n in range(len(M)):
    max_sum = 0
    max_rad = 0
    for i in range(len(M[n])):
        c_sum = 0
        for j in range(len(M[n])):
            if j != i:
                c_sum += abs(M[n][j][i])
        rad = M[n][j][i]
        
        #plot circle
        axis[n].set_aspect( 1 )
        axis[n].add_artist( plt.Circle(( c_sum, 0 ), rad, fill = False  ))
        axis[n].plot(c_sum, 0, 'ro')
        
        #set max values for limits
        if c_sum > max_sum:
            max_sum = c_sum
        if rad > max_rad:
            max_rad = rad
        
    axis[n].set_xlim(-max_sum-max_rad,max_sum+max_rad)
    axis[n].set_ylim(-max_rad,max_rad)
    axis[n].set_title(T[n]+" by column")
plt.show()
        
```

```python
#Problem 7. BUID
import numpy as np
from numpy import tril
from numpy import triu
from numpy.linalg import det
from numpy.linalg import eig
from numpy.linalg import matrix_rank
from numpy.linalg import qr


A = np.array([[1,7,5],[4,2,5],[8,1,8]])

w,v = eig(A)

L = tril(A)
U = triu(A)

Q, R = qr(A, 'reduced')

QT = np.transpose(Q)

print("A determinant = ", round(det(A),2))
print("A rank = ", matrix_rank(A))
print("A eigenvalues = ", w)
print("A eigenvectors = ", v)
print("L = ", L)
print("U = ", U)
print("Q = ", Q)
print("R = ", R)
print("QQT = ", Q.dot(QT))
print("Q is not orthogonal")
```
