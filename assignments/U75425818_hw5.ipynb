```python
#Problem 3. Compute Vi norm and Ev
import numpy as np
from numpy.linalg import norm
import matplotlib.pyplot as plt

V1 = np.array([1,0])
V2 = np.array([5,5])
V3 = np.array([-4,4])

V1_norm = norm(V1)
V2_norm = norm(V2)
V3_norm = norm(V3)

print("V1 norm = ", round(V1_norm,2))
print("V2 norm = ", round(V2_norm,2))
print("V3 norm = ", round(V3_norm,2))

ev1 = V1/V1_norm
ev2 = V2/V2_norm
ev3 = V3/V3_norm

  
# Creating plot
fig = plt.figure(figsize=(5, 5)) 
ax = fig.add_subplot(111)
ax.grid(alpha=1) 
ax.set(xlim=(-4, 5), ylim=(0, 5))
ax.annotate('', xy=V1, xytext=(0, 0), arrowprops=dict(facecolor='g'))
ax.annotate('', xy=ev1, xytext=(0, 0), arrowprops=dict(facecolor='r'))
ax.annotate('', xy=V2, xytext=(0, 0), arrowprops=dict(facecolor='g'))
ax.annotate('', xy=ev2, xytext=(0, 0), arrowprops=dict(facecolor='r'))
ax.annotate('', xy=V3, xytext=(0, 0), arrowprops=dict(facecolor='g'))
ax.annotate('', xy=ev3, xytext=(0, 0), arrowprops=dict(facecolor='r'))
plt.title('Problem 3')
  

# Show plot with grid
plt.grid()
plt.show()
```
