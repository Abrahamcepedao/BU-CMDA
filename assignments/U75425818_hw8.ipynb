```python
#Problem 1.
import numpy as np
from numpy.linalg import eig

A = np.array([[1,2],[3,4]])
I = np.array([[1,0],[0,1]])

w,v = eig(A)

print("A eigenvalues = ", w)
print("A eigenvectors = ", v)

A2 = A.dot(A)
A5N = -5*A
print(A2)
print(A5N)

print("P(A) = ",A2+A5N-2*I)

```

```python
#Problem 2.
import numpy as np
from numpy.linalg import eig

A = np.array([[1,0,0],[0,0,1],[0,1,0]])
B = np.array([[1,1,1],[0,1,1],[0,0,1]])
I = np.array([[1,0,0],[0,1,0],[0,0,1]])

w,v = eig(A)
w2,v2 = eig(B)

print("A eigenvalues = ", w)
print("A eigenvectors = ", v)
print("B eigenvalues = ", w2)
print("B eigenvectors = ", v2)

#P(A) = -A^3 + A + A^2 -1I
A2 = A.dot(A)
A3N = -1* A2.dot(A)
IN = -1*I

print("P(A) = ", A3N + A + A2 + IN)

#(-x+1)(-x+1)(-x+1) = (x^2 -2x + 1)(-x+1) = (-x^3 + 3x^2 - 3x + 1)
#P(B) = -B^3 + 3B^2 - 3B -1I
B2 = B.dot(B)
B2N = 3*B2
B3N = -1*B2.dot(B)
BN = -3*B

print("BN = ", BN)

print("P(B) = ", B3N + B2N + BN + I)
```

```python
#Problem 3.
import numpy as np
from numpy.linalg import eig
import matplotlib.pyplot as plt

V1 = np.array([5,8])
V2 = np.array([1,8])

#Step 1 - U1 = V1
U1 = V1

fig = plt.figure(figsize=(5, 5)) 
ax = fig.add_subplot(111)
ax.grid(alpha=1) 
ax.set(xlim=(0, 5), ylim=(0, 8))
ax.annotate('', xy=V1, xytext=(0, 0), arrowprops=dict(facecolor='g'))
ax.annotate('', xy=V2, xytext=(0, 0), arrowprops=dict(facecolor='r'))
ax.set_title("Step 1")
plt.show()

#Step 2 - Projection U1 (V1)
AB = V1[0]*V2[0] + V1[1]*V2[1]
DA = np.sqrt(V1[0]*V1[0] + V1[1]*V1[1])
DB = np.sqrt(V2[0]*V2[0] + V2[1]*V2[1])
cos_theta = AB / (DA*DB)

PU1 = cos_theta * V1 #ProjU1(V1)

fig = plt.figure(figsize=(5, 5)) 
ax = fig.add_subplot(111)
ax.grid(alpha=1) 
ax.set(xlim=(0, 5), ylim=(0, 8))
ax.annotate('', xy=V1, xytext=(0, 0), arrowprops=dict(facecolor='g'))
ax.annotate('', xy=V2, xytext=(0, 0), arrowprops=dict(facecolor='r'))
ax.annotate('', xy=PU1, xytext=(0, 0), arrowprops=dict(facecolor='b'))
ax.set_title("Step 2")
plt.show()

#Step 3 - Projection U1 (V1) and E1
E1 = U1 / DA
fig = plt.figure(figsize=(5, 5)) 
ax = fig.add_subplot(111)
ax.grid(alpha=1) 
ax.set(xlim=(0, 5), ylim=(0, 8))
ax.annotate('', xy=V1, xytext=(0, 0), arrowprops=dict(facecolor='g'))
ax.annotate('', xy=V2, xytext=(0, 0), arrowprops=dict(facecolor='r'))
ax.annotate('', xy=PU1, xytext=(0, 0), arrowprops=dict(facecolor='b'))
ax.annotate('', xy=E1, xytext=(0, 0), arrowprops=dict(facecolor='black'))
ax.set_title("Step 3")
plt.show()

#Step 4 - Projection U2 (V2) and E2
U2 = V2 - PU1 #U1
DU2 = np.sqrt(U2[0]*U2[0] + U2[1]*U2[1]) #|U2|
E2 = U2 / DU2 #E2

fig = plt.figure(figsize=(5, 5)) 
ax = fig.add_subplot(111)
ax.grid(alpha=1) 
ax.set(xlim=(-4, 5), ylim=(0, 8))
ax.annotate('', xy=V1, xytext=(0, 0), arrowprops=dict(facecolor='g'))
ax.annotate('', xy=V2, xytext=(0, 0), arrowprops=dict(facecolor='r'))
ax.annotate('', xy=PU1, xytext=(0, 0), arrowprops=dict(facecolor='b'))
ax.annotate('', xy=E1, xytext=(0, 0), arrowprops=dict(facecolor='black'))
ax.annotate('', xy=U2, xytext=(0, 0), arrowprops=dict(facecolor='c'))
ax.annotate('', xy=E2, xytext=(0, 0), arrowprops=dict(facecolor='black'))
ax.set_title("Step 4")
plt.show()
```
