```python
#calculate days since long count
from datetime import date

def days_since_long_count(lst):
    #long_count is a string in the format "
    base = [144000,7200,360,20,1] #base of each long count unit 
    days = 0 

    for i in range(0,len(lst)):
        days = days + (base[i] * lst[i])

    return days

def days_to_long_count(days): 
    lst = []          
    base=[20,18,20,20,20] #base of each long count unit 
    i = 0    
    rem = 0    
    while days > 0:
        rem = days % base[i]
        lst.append(rem)
        days = days//base[i]
        i = i + 1
    lst.reverse()
    return lst

def habb_to_gregorian(habb):
    #habb is a string in the format: 5 Wayeb "
    habb = habb.split(" ")
    day = int(habb[0])
    month = habb[1]

    habb_monhts = ["Pohp", "Wo", "Sip", "Sots", "Sek", "Xul", "Yaxkin", "Mol", "Chen", "Yax", "Sak", "Keh", "Mak", "Kankin", "Muwan", "Pax", "Kayab", "Kumku", "Wayeb"]
    gregorian_months = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]
    days_in_gregorian_month = [31,28,31,30,31,30,31,31,30,31,30,31]

    mm = habb_monhts.index(month)
    num_days = mm * 20 + day

    #convert to gregorian
    i = 0
    while num_days > days_in_gregorian_month[i]:
        num_days = num_days - days_in_gregorian_month[i]
        i = i + 1

    return num_days,gregorian_months[i]

def gregorian_to_habb(day,month):
    habb_monhts = ["Pohp", "Wo", "Sip", "Sots", "Sek", "Xul", "Yaxkin", "Mol", "Chen", "Yax", "Sak", "Keh", "Mak", "Kankin", "Muwan", "Pax", "Kayab", "Kumku", "Wayeb"]
    gregorian_months = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]
    days_in_gregorian_month = [31,28,31,30,31,30,31,31,30,31,30,31]
    num_days = 0

    for i in range(0,gregorian_months.index(month)):
        num_days = num_days + days_in_gregorian_month[i]

    num_days = num_days + day

    habb_month = num_days//20
    habb_day = num_days%20

    return habb_day,habb_monhts[habb_month]

def long_count_to_calendar_round(long_count):
    habb_monhts = ["Pohp", "Wo", "Sip", "Sots", "Sek", "Xul", "Yaxkin", "Mol", "Chen", "Yax", "Sak", "Keh", "Mak", "Kankin", "Muwan", "Pax", "Kayab", "Kumku", "Wayeb"]
    tolzkin_months = ["Imix", "Ik", "Akbal", "Kan", "Chicchan", "Cimi", "Manik", "Lamat", "Muluk", "Ok", "Chuen", "Eb", "Ben", "Ix", "Men", "Cib", "Caban", "Etznab", "Cauac", "Ahau"]
    #long_count is a string in the format "
    long_count = long_count.split(".")
    long_count = [int(i) for i in long_count]
    long_count.reverse()  # Invert date order

    #veintena
    v = long_count[0]-1

    #trecena
    mt = [1,-6,-4,-2,-1]
    tr = (4 + ((long_count[0] * mt[0]) + (long_count[1] * mt[1]) + (long_count[2] * mt[2]) + (long_count[3] * mt[3]) + (long_count[4] * mt[4]))) % 13

    #haab
    mh = [1, 20, -5, -100, -175]
    hb = (-17 + ((long_count[0] * mh[0]) + (long_count[1] * mh[1]) + (long_count[2] * mh[2]) + (long_count[3] * mh[3]) + (long_count[4] * mh[4]))) % 365
    habb_month = hb//20
    habb_day = hb%20

    return tr, tolzkin_months[v], habb_day, habb_monhts[habb_month]

def numOfDays(date1, date2):
    return (date1-date2).days

def gregorian_to_mayan_date(day,month,year):
    init_date = date(3114, 8, 13)
    final_date = date(year, month, day)
    zero_date = date(1, 1, 1)
    days = numOfDays(init_date, zero_date)
    days += numOfDays(final_date, zero_date)-80
    lst = days_to_long_count(days)
    long_count = ".".join(str(i) for i in lst)
    
    print(long_count, long_count_to_calendar_round(long_count))


print(days_since_long_count([9,8,9,0,0]))
print(days_to_long_count(1746840))
print(habb_to_gregorian("5 Yaxkin"))
print(gregorian_to_habb(31,"December"))
print(long_count_to_calendar_round("12.19.5.5.0"))
gregorian_to_mayan_date(6,12,2022)
```
